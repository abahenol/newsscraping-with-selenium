# News Portal Scraping - beritajatim.com
With selenium package

```python
#import package
from time import sleep, strftime
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
#use Chrome Driver as Web Driver
chromedriver_path = 'C:/Program Files/Chrome Driver/chromedriver.exe'
driver = webdriver.Chrome(executable_path=chromedriver_path) # This will open the Chrome window
sleep(2)
```

```python
web_berita = ('https://beritajatim.com/kanal/hukum-kriminal/')
driver.get(web_berita)
```

## Main Scraping Process

```python
data=[]
#looping for news portal page
#change the range according with the needs
for a in range(1,3): 
    web_berita = ('https://beritajatim.com/kanal/hukum-kriminal/page/'+str(a))
    driver.get(web_berita)
    sleep(2)
    a = driver.find_elements_by_tag_name("h3")[0:10]
    text_a = []
    #looping for news headline in a page
    for k in a: 
        l = k.text
        text_a.append(l)
    #looping for get all the news content
    for m in range(0,10):
        driver.find_element_by_link_text(text_a[m]).click()
        try :
            sleep(2)
            title = driver.find_element_by_tag_name("h1").text
            time = driver.find_element_by_tag_name("time").text
            place = driver.find_element_by_class_name("entry-content").text[:20]
            place_list = place.split()
            place = place_list[0]
            isi_berita = driver.find_element_by_class_name("entry-content").text[:1000]
            data.append([title, time, place, isi_berita])
            driver.get(web_berita)
        except NoSuchElementException :
            driver.get(web_berita)
```

```python
#take the list into dataframe
pd.DataFrame(data, columns=["judul", "waktu", "tempat", "isi_berita"])
```
